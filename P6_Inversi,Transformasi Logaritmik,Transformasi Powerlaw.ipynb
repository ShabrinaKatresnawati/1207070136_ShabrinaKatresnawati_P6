```python
#import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#membaca file gambar
img = cv2.imread("matcha.jpg")
```

```python
#mendapatkan/define resolusi dan tipe gambar
img_height=img.shape[0]#pengambilan nilai tinggi citra 
img_width = img.shape[1]#pengambilan nilai lebar citra
img_channel= img.shape[2]#pengambilan jumlah channel citra 
```

```python
#membuat variable img_inversi untuk menampung
img_inversi = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi inversi grayscale dengan nilai yang menjadi parameter 
def inversi_grayscale(nilai):
    for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
        for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
            #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = nilai - gray
            #piksel yang sedang diakses pada baris y dan kolom x.
            img_inversi[y][x] = (gray, gray, gray)
```

```python
#membuat fungsi inversi rgb dengan nilai yang menjadi parameter
def inversi_rgb(nilai):
    for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
        for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
            #mengambil nilai RGB untuk setiap piksel gambar pada posisi (y, x). untuk masing-masing komponen warna (merah, hijau, dan biru), program menghitung nilai yang baru sebagai perbedaan antara nilai masukan dan nilai yang diambil dari gambar
            red = img[y][x][0]
            red = nilai - red
            green = img[y][x][1]
            green = nilai - green
            blue = img[y][x][2]
            blue = nilai - blue
            #piksel yang sedang diakses pada baris y dan kolom x.
            img_inversi[y][x] = (red, green, blue)
```

```python
#menampilkan inversi grayscale
inversi_grayscale(255)
plt.imshow(img_inversi)
plt.title("Inversi Grayscale")
plt.show()
#menampilkan inversi rgb
inversi_rgb(255)
plt.imshow(img_inversi)
plt.title("Inversi RGB")
plt.show()
```

```python
#membuat variable img_log untuk menampung
img_log = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi log dengan nilai c yang menjadi parameter
def log(c):
    for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
        for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
            #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
            gray = (int(red) + int(green) + int(blue)) / 3
            #menghasilkan citra grayscale baru dengan nilai piksel yang dihitung menggunakan rumus log transform 
            gray = int(c * np.log(gray + 1))
            if gray > 255:#jika nilai keabuan (gray) lebih besar dari 225 maka nilai keabuan diatur ke nilai maksimum yang dapat direpresentasikan oleh tipe data uint8, yaitu 255.
                gray=255
            #nilai keabuan (gray) lebih kecil dari 0. Jika ya, maka nilai keabuan diatur ke 0
            if gray < 0:
                gray = 0
            #piksel yang sedang diakses pada baris y dan kolom x.
            img_log[y][x] = (gray, gray, gray)
```

```python
# menghitung nilai logaritma natural dari angka 30
log(30)
#menampilkan gambar
plt.imshow(img_log)
plt.title("Log")
plt.show()
```

```python
#Membuat variabel img_nthpower untuk menampung hasil
img_nthpower = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi nthpower dengan nilai c,y yang menjadi parameter
def nthpower(c, y):
    #nilai dari parameter c,y akan dibagi dengan 100 
    thc = c / 100
    thy = y / 100
    for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
        for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
            #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(thc * pow(gray, thy))
            #jika nilai keabuan (gray) lebih besar dari 225 maka nilai keabuan diatur ke nilai maksimum yang dapat direpresentasikan oleh tipe data uint8, yaitu 255.
            if gray > 225:
                gray=255
            #nilai keabuan (gray) lebih kecil dari 0. Jika ya, maka nilai keabuan diatur ke 0
            if gray < 0:
                gray = 0
            #piksel yang sedang diakses pada baris y dan kolom x.
            img_nthpower[y][x] = (gray, gray, gray)
```

```python
#memasukan nilai parameter c,y
nthpower(50, 100)
#menampilkan gambar
plt.imshow(img_nthpower)
plt.title("Nth Power")
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("matcha.jpg")
```

```python
#mendapatkan/define resolusi dan tipe gambar
img_height=img.shape[0]#pengambilan nilai tinggi citra 
img_width = img.shape[1]#pengambilan nilai lebar citra
img_channel= img.shape[2]#pengambilan jumlah channel citra 
```

```python
#membuat variable img_grayscale untuk menampung
img_grayscale = np.zeros(img.shape, dtype=np.uint8)

for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
    for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
        #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
        gray = (int(red) + int(green) + int(blue)) / 3
        #piksel yang sedang diakses pada baris y dan kolom x.
        img_grayscale[y][x] = (gray, gray, gray)
#menampilkan citra dari img_grayscale      
plt.imshow(img_grayscale)
plt.title("Grayscale")
plt.show()
```

```python
#pembuatan array dengan Numpy yang berukuran 256x1
hg = np.zeros((256))
```

```python
#looping pada setiap piksel gambar untuk merekam kemunculan nilai intensitasnya
for x in range(0, 256):
    hg[x] = 0
```

```python
for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
    for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
        #pengambilan nilai intensitas gray dari piksel gambar yang sedang diproses
        gray = img_grayscale[y][x][0]
        #nilai intensitas gray diincrement dengan 1
        hg[gray] += 1
```

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()
#menentukan interval nilai intensitas yang akan ditampilkan dalam histogram.
bins = np.linspace(0, 256, 100)
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```

```python
#Membuat variabel untuk menyimpan data gambar. hgr, hgg, dan hgb untuk histogram warna merah, hijau, dan biru, sedangkan hgrgb untuk histogram gabungan ketiga warna.
hgr = np.zeros((256))
hgg = np.zeros((256))
hgb = np.zeros((256))
hgrgb = np.zeros((768))
```

```python
#Menginisialisasi semua nilai histogram dengan 0
for x in range(0, 256):
    hgr[x] = 0
    hgg[x] = 0
    hgb[x] = 0
    
for x in range(0, 768):
    hgrgb[x] = 0
```

```python
#Melakukan iterasi pada setiap piksel gambar dan menghitung frekuensi kemunculan tiap nilai warna pada setiap kanal warna (merah, hijau, biru)
for x in range(0, 256):
    hgr[x] = 0
    hgg[x] = 0
    hgb[x] = 0
    
for x in range(0, 768):
    hgrgb[x] = 0

# th = int(256/64)
temp = [0]
#perulangan untuk mengakses setiap baris pada gambar.
for y in range(0, img.shape[0]):
    for x in range(0, img.shape[1]):
        #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses  
        red = int(img[y][x][0])
        green = int(img[y][x][1])
        blue = int(img[y][x][2])
        green = green + 256
        blue = blue + 512
        #Melakukan iterasi pada setiap piksel gambar dan menghitung frekuensi kemunculan tiap nilai warna pada setiap kanal warna (merah, hijau, biru)
        hgrgb[red] += 1
        hgrgb[green] += 1
        hgrgb[blue] += 1
#menentukan interval nilai intensitas yang akan ditampilkan dalam histogram.
binsrgb = np.linspace(0, 768, 100)
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(hgrgb, binsrgb, color="black", alpha=0.5)
# plt.plot(hgrgb)
plt.title("Histogram Red Green Blue")
plt.show()
```

```python
for y in range(0, img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
    for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
        #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        #Melakukan iterasi pada setiap piksel gambar dan menghitung frekuensi kemunculan tiap nilai warna pada setiap kanal warna (merah, hijau, biru)
        hgr[red] += 1
        hgg[green] += 1
        hgb[blue] += 1
#menentukan interval nilai intensitas yang akan ditampilkan dalam histogram.
bins = np.linspace(0, 256, 100)
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(hgr, bins, color="red", alpha=0.5)
plt.title("Histogram Red")
plt.show()
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(hgg, bins, color="green", alpha=0.5)
plt.title("Histogram Green")
plt.show()
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(hgb, bins, color="blue", alpha=0.5)
plt.title("Histogram Blue")
plt.show()
```

```python
# menginisialisasi nilai pada array
hgk = np.zeros((256))
c = np.zeros((256))
#looping pada setiap piksel gambar untuk merekam kemunculan nilai intensitasnya
for x in range(0, 256):
    hgk[x] = 0
    c[x] = 0
#menghitung nilai histogram untuk setiap level keabuan dalam citra grayscale
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgk[gray] += 1
                
c[0] = hgk[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgk[x]
#nilai maksimum dari kumulatif histogram c
hmaxk = c[255]
#Melakukan normalisasi pada setiap nilai c dengan mengalikan nilai tersebut dengan 190 dan membaginya dengan hmaxk
for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Kumulatif")
plt.show()
```

```python
# menginisialisasi nilai pada array
hgh = np.zeros((256))
h = np.zeros((256))
c = np.zeros((256))
#looping pada setiap piksel gambar untuk merekam kemunculan nilai intensitasnya
for x in range(0, 256):
    hgh[x] = 0
    h[x] = 0
    c[x] = 0

for y in range(0, img_height):
    for x in range(0, img_width):
        #mengambil nilai grayscale dan menambahkan 1 ke elemen "hgk"
        gray = img_grayscale[y][x][0]
        hgh[gray] += 1
#Menghitung nilai kumulatif              
h[0] = hgh[0]
for x in range(1, 256):
     h[x] = h[x-1] + hgh[x]
#Normalisasi
for x in range(0, 256):
     h[x] = h[x] / img_height / img_width
#Menghitung histogram equalization
for x in range(0, 256):
    hgh[x] = 0
    
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        gray = h[gray] * 255
        hgh[int(gray)] += 1
#Menghitung nilai kumulatif
c[0] = hgh[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgh[x]

#nilai maksimum dari kumulatif histogram c
hmaxk = c[255]
#Melakukan normalisasi pada setiap nilai c dengan mengalikan nilai tersebut dengan 190 dan membaginya dengan hmaxk
for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk
#menampilkan histogram memasukkan array hg dan bins sebagai parameter, serta menentukan warna dan transparansi histogram dengan color dan alpha
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Hequalisasi")
plt.show()
```
